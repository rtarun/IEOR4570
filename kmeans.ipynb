```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
```

sampling points from sphere

```python
#% matplotlib inline

def sampleBallPoints(center, radius):
    
    R = radius
    
    center_x = center[0]
    center_y = center[1]
    center_z = center[2]
    
    x = []
    y = []
    z = []

    for i in range(50):

        phi = random.uniform(0,2*math.pi)
        costheta = random.uniform(-1,1)
        u = random.random()

        theta = math.acos(costheta)

        r = R*(u**(1/3))

        x.append((r * math.sin( theta) * math.cos( phi ))+center_x)
        y.append((r * math.sin( theta) * math.sin( phi ))+center_y)
        z.append((r * math.cos( theta ))+center_z)

   
    #Axes3D.scatter(x,y,z)
    return x, y, z
```

```python
radius = 10

center1 = [0,0,0]
center2 = [0,20,0]

x1, y1, z1 = sampleBallPoints(center1,radius)
x2, y2, z2 = sampleBallPoints(center2,radius)

sphere1 = list(zip(x1, y1, z1))
sphere2 = list(zip(x2, y2, z2))

data = sphere1+sphere2
```

Running K Means on these two spheres

iterations to obtain centers B(c1,9) and B(c2,9)

```python
from sklearn import cluster

k = 2

for i in range(1,1000,100):
    kmeans = cluster.KMeans(n_clusters=k, max_iter = i, init = 'random')
    kmeans.fit(data)

#labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    
    print("Iteration:",i)
    print(centroids)
```

Cost function as function of iterations

```python
for i in range(1,100):
    kmeans = cluster.KMeans(n_clusters=k, max_iter = i, init = 'random')
    kmeans.fit(data)

#labels = kmeans.labels_
    costFunction = kmeans.inertia_
    
    print("Iteration:",i)
    print(costFunction)
```
